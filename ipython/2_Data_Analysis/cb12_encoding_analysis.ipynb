```python
# set seed to reproduce results with tf and keras
from numpy.random import seed
seed(1)
# other imports
import numpy as np
import matplotlib.pyplot as plt
import pickle
import numpy as np
import pandas as pd
import pandas as pd

path =  "../../data/"
dataset = "cb12/"

raw_path = path + dataset + "raw/" 
interim_path = path + dataset + "interim/"
processed_path = path + dataset + "processed/"
```

```python
common_items = pd.read_csv(interim_path + "common_items.csv", header=0, sep='\t')
common_items.head()
```

```python
common_items[common_items.columns[2:-1]].max().max()
```

```python
common_items[common_items.columns[2:-1]].min().min()
```

```python
content_fields = ["state"]

one_hot = pd.get_dummies(data=common_items, columns=content_fields)

print(one_hot.shape)

one_hot = one_hot.set_index("item_id")
one_hot.head()
```

```python
one_hot.drop(columns=['session_id'], inplace=True)
```

```python
combined = one_hot.applymap(str)
combined["repr"] = combined.apply(lambda x: ''.join(x), axis=1)
combined.head()
```

```python
unique = combined["repr"].nunique()
total = len(combined["repr"])
ratio = unique / total
print(unique, total, ratio)
```
